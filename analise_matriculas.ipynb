# Criando a primeira base de dados: ntodos

## Importando 

```python
# importando o pandas:
import pandas as pd

# importando as duas bases de dados: nomes femininos e masculinos em DataFrames (DF)
nf = pd.read_json('https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=20&sexo=f')  # nomes femininos
nm = pd.read_json('https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=20&sexo=m')  # nomes masculinos
```

```python
# exibindo uma amostra do DF:
nf.sample(5)
```

```python
# exibe a soma dos registros dos dois DFs:
print('Quantidade de nomes femininos e masculinos:', len(nf) + len(nm))
```

## Criando o DataFrame ntodos

```python
# criando o DF principal, concatenando os dois DFs existentes:
ntodos = pd.concat([nf, nm])['nome'].to_frame()

# exibindo o novo DF criado:
ntodos
```

## Incluindo ID dos alunos

```python
# importando o NumPy
import numpy as np

# definindo a seed com np.random.seed()
np.random.seed(124)
```

```python
# criando um novo campo id_aluno e definindo valores aleatórios de acordo com a len(ntodos), com np.random.permutation()
ntodos['id_aluno'] = np.random.permutation(len(ntodos)) + 1
ntodos.sample(3)
```

## Criando os campos 'domino' e 'email'

```python
# lista com os domínios padrão:
doms = ['@dominiodeemail.com.br', '@outrodominio.com.br']

# criando o novo campo 'dominio':
ntodos['dominio'] = np.random.choice(doms, len(ntodos))

# criando o novo campo 'email':
ntodos['email'] = ntodos['nome'].str.lower() + ntodos['dominio']

# consertando o index
ntodos.index = range(len(ntodos))
ntodos
```

# Criando a segunda base de dados: cursos

```python
import html5lib
```

```python
# importando do URL para uma lista:
cursos_lista = pd.read_html('http://tabela-cursos.herokuapp.com/index.html')

# transformando em um DF:
cursos = cursos_lista[0]

# renomeando a coluna para uma melhor manipulação e organização:
cursos = cursos.rename(columns={'Nome do curso': 'nome_do_curso'})

cursos.head()
```

```python
# criando uma nova coluna 'id':
cursos['id'] = cursos.index + 1

cursos.head()
```

```python
# definindo essa nova coluna como o index do DF com set_index():
cursos = cursos.set_index('id')
cursos.head()
```

# Criando a nova coluna 'matriculas'

```python
# relembrando o DF ntodos:
ntodos.sample(3)
```

```python
# criando a nova coluna 'matriculas':
ntodos['matriculas'] = np.ceil(np.random.exponential(size=len(ntodos)) * 1.5).astype(int)
ntodos.head()
```

## Criando o novo DF 'matriculas'

```python
# relembrando o ntodos:
ntodos.sample(3)
```

```python
x = np.random.rand(20)  # gera um array com 20 números, de 0 a 1, aleatoriamente
prob = x / sum(x)  # divide cada elemento do 'x' por sum(x)

print(f'VISUALIZAR VARIÁVEIS:\n\nx = {x}. \n\nprob = {prob}. \n\nsum(x) = {sum(x)}')
```

```python
# criando um novo DF 'matriculas':

todas_matriculas = []  # lista vazia que será preenchida a cada iteração do for abaixo

for i, r in ntodos.iterrows():  # itera no DF ntodos. i = index; r = row(pd.Series)
    id = r.id_aluno  # localiza e armazena o valor da coluna 'id_aluno' da linha atual
    qtde_matriculas = r.matriculas  # localiza e armazena o valor da coluna 'matriculas' da linha atual.
    for i in range(qtde_matriculas):  # itera sobre o número (int) que contém na qtde_matriculas
        insert_mat = [id, np.random.choice(cursos.index, p=prob)]  
        # ^ cria uma lista com o 'id' e um 'id' aleatório dentre os ids do DF 'cursos'
        todas_matriculas.append(insert_mat)  # append a lista criada acima na lista 'todas_matriculas'
        
matriculas = pd.DataFrame(todas_matriculas, columns=['id_aluno', 'id_curso'])  
# ^ cria o novo DF conforme a lista 'todas_matriculas' criada acima

matriculas.head()
```

## Criando o novo DF 'matriculas_por_curso'

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'id_aluno': 'qtde_de_alunos'})
matriculas_por_curso
```

# Criando um BD SQLite com SQLAlchemy

```python
from sqlalchemy import create_engine, MetaData, Table, inspect # adicionando o método inspect

# criando a engine sqlite (na memória)
engine = create_engine('sqlite://')

# transformando o DF matriculas_por_curso em sql:
matriculas_por_curso.to_sql('matriculas_bd', engine)

inspector = inspect(engine) # criando um Inspector object
print(inspector.get_table_names()) # Exibindo as tabelas com o inspector
```

```python
# verificando o type da var engine criada
type(engine)
```

## Queries no banco de dados

```python
query = "select * from matriculas_bd where qtde_de_alunos <= 6"  # variável com a query
pd.read_sql(query, engine, index_col='id_curso')
```

### Criando um novo DF 'muitas_matriculas'

```python
# criando o novo DF a partir do sqlite, alterando a sequência das colunas no final
muitas_matriculas = pd.read_sql_table('matriculas_bd', engine, index_col='id_curso', columns=['nome_do_curso', 'qtde_de_alunos'])
```

```python
muitas_matriculas.head()
# ou seja, esse novo DF muitas_matriculas, NO MOMENTO, está idêntico ao DF matriculas_por_curso.
```

```python
# atualizando a muitas_matriculas para conter apenas os valores da query abaixo
muitas_matriculas = muitas_matriculas.query('qtde_de_alunos >= 7')  # apenas os elementos 'qtde_de_alunos >= 7'

# exibindo:
muitas_matriculas
```

## Escrevendo no banco de dados
